```python
import numpy as np
import matplotlib.pyplot as plt

learning_rates = [0.001, 0.01, 0.1, 0.5]
steps = 50

for lr in learning_rates:
    x = 4
    x_history = [x]
    
    for i in range(steps):
        grad = 2*x
        x = x - lr *grad
        x_history.append(x)
    
    plt.plot(x_history, label=f'lr={lr}')

plt.title("Effect of Learning Rate")
plt.xlabel("Steps")
plt.ylabel("x value")
plt.legend()
plt.grid(True)
plt.show()
```
