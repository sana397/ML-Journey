```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

i = Image.open(r"C:\Users\anjan\ML-Journey\sky_for_ml.jpg").convert("L")
i = i.resize((800, 600))  # ADD THIS LINE
ia = np.array(i)

brighter = np.clip(ia * 1.5, 0, 255).astype(np.uint8)
darker = np.clip(ia * 0.5, 0, 255).astype(np.uint8)

plt.imshow(ia, cmap='gray')
plt.title("Original")
plt.axis('off')
plt.show()

plt.imshow(brighter, cmap='gray')
plt.title("Brighter")
plt.axis('off')
plt.show()

plt.imshow(darker, cmap='gray')
plt.title("Darker")
plt.axis('off')
plt.show()

```
