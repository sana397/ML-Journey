```python
import numpy as np
import matplotlib.pyplot as plt
x = 4
y = 4
learning_rate = 0.1
steps = 20
x_history = [x]
y_history = [y]
for i in range(steps):
    grad_x = 2*x
    grad_y = 2*y
    x = x - learning_rate * grad_x
    y = y - learning_rate * grad_y
    x_history.append(x)
    y_history.append(y)
plt.plot(x_history, label='x')
plt.plot(y_history, label='y')
plt.title("Gradient Descent on f(x,y) = x² + y²")
plt.xlabel("Step")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
print("Final x:", x)
print("Final y:", y)
```
