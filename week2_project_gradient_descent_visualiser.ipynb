```python

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

x_pos = 4
y_pos = 4
learning_rate = 0.1
steps = 20

path_x = [x_pos]
path_y = [y_pos]
path_z = [x_pos**2 + y_pos**2]

for i in range(steps):
    grad_x = 2 * x_pos
    grad_y = 2 * y_pos
    x_pos = x_pos - learning_rate * grad_x
    y_pos = y_pos - learning_rate * grad_y
    path_z.append(x_pos**2 + y_pos**2)
    path_x.append(x_pos)
    path_y.append(y_pos)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.6)
ax.plot(path_x, path_y, path_z, 'r-o', markersize=4)
ax.set_title('Gradient Descent on 3D Surface')
plt.show()

learning_rates = [0.001, 0.01, 0.1, 0.5]
steps = 50

for lr in learning_rates:
    x = 4
    x_history = [x]
    for i in range(steps):
        grad = 2 * x
        x = x - lr * grad
        x_history.append(x)
    plt.plot(x_history, label=f'lr={lr}')

plt.title("Effect of Learning Rate")
plt.xlabel("Steps")
plt.ylabel("x value")
plt.legend()
plt.grid(True)
plt.show()

```
