```python
import numpy as np
import matplotlib.pyplot as plt
x = 4
learning_rate = 0.1
steps = 20
x_history = [x]
for i in range(steps):
    gradient = 2 * x
    x = x-learning_rate *gradient
    x_history.append(x)
plt.plot(x_history)
plt.title("gradient descent")
plt.xlabel("step")
plt.ylabel("x value")
plt.grid(True)
plt.show()
print("final x:",x)
```
