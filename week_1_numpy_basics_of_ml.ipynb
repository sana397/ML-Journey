```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy.ndimage import rotate
# 1. Load as NumPy matrix
img = np.array(Image.open("photo.jpg").convert("L")) # grayscale uint8
print("Shape:", img.shape, "| dtype:", img.dtype)
# 2. Brightness via scalar multiply + clip
bright = np.clip(img * 1.5, 0, 255).astype(np.uint8) # +50%
dark = np.clip(img * 0.5, 0, 255).astype(np.uint8) # -50%
# 3. Rotation transformation
rotated = rotate(img, angle=45, reshape=False)
# 4. Side-by-side comparison
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
for ax, im, t in zip(axes,
[img, bright, dark, rotated],
["Original", "+50% Bright", "-50% Dark", "Rotated 45"]):
    ax.imshow(im, cmap="gray")
ax.set_title(t)
ax.axis("off")
plt.tight_layout()
plt.savefig("output.png", dpi=150)
plt.show()
```
