```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

i = Image.open(r"C:\Users\anjan\ML-Journey\sky_for_ml.jpg").convert("L")
i = i.resize((800, 600))
ia = np.array(i)

brighter = np.clip(ia * 1.5, 0, 255).astype(np.uint8)
darker = np.clip(ia * 0.5, 0, 255).astype(np.uint8)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(ia, cmap='gray')
axes[0].set_title("Original")
axes[0].axis('off')
axes[1].imshow(brighter, cmap='gray')
axes[1].set_title("Brighter")
axes[1].axis('off')
axes[2].imshow(darker, cmap='gray')
axes[2].set_title("Darker")
axes[2].axis('off')
plt.show()
```
