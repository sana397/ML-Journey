```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt

# 1. DEFINE FILE PATHS
# Using 'r' before the string handles the backslashes in Windows paths correctly
flair_path = r"C:\Users\anjan\ML-Journey\MRI-Project\data\BraTS2020_TrainingData\MICCAI_BraTS2020_TrainingData\BraTS20_Training_001\BraTS20_Training_001_flair.nii"
seg_path = r"C:\Users\anjan\ML-Journey\MRI-Project\data\BraTS2020_TrainingData\MICCAI_BraTS2020_TrainingData\BraTS20_Training_001\BraTS20_Training_001_seg.nii"

# 2. LOAD THE IMAGES
flair_img = nib.load(flair_path)
flair_data = flair_img.get_fdata()

seg_img = nib.load(seg_path)
seg_data = seg_img.get_fdata()

# 3. PRINT DATA INFO (For Debugging)
print(f"FLAIR Shape: {flair_data.shape}")
print(f"Segmentation Shape: {seg_data.shape}")
print(f"Unique Labels in Segmentation: {np.unique(seg_data)}") 
# BraTS labels: 0=Background, 1=Necrotic Core, 2=Edema, 4=Enhancing Tumor

# 4. VISUALIZATION
slice_idx = 77  # The slice you used previously

plt.figure(figsize=(12, 6))

# Subplot 1: The Raw MRI Scan (Input)
plt.subplot(1, 2, 1)
plt.imshow(flair_data[:, :, slice_idx], cmap='gray')
plt.title(f'FLAIR Scan (Slice {slice_idx})')
plt.axis('off')

# Subplot 2: The Ground Truth (Target/Labels)
# We use 'jet' or 'viridis' colormap to make the different tumor parts colorful
plt.subplot(1, 2, 2)
plt.imshow(seg_data[:, :, slice_idx], cmap='jet')
plt.title('Doctor\'s Segmentation Mask')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
p_spam = 0.4
p_normal= 0.6
p_free_given_spam = 0.8
p_free_given_normal = 0.1
p_free = (p_free_given_spam * p_spam) + (p_free_given_normal * p_normal)
p_spam_given_free = (p_free_given_spam*p_spam)/p_free
print("p(free):",round(p_free,2))
print("p(spam|free):",round(p_spam_given_free,2))
print("conclusion:if email free,probability tis is spam=",str(round(p_spam_given_free *100,1)) + "%")
```

```python
# Visualise Bayes update
import matplotlib.pyplot as plt
import numpy as np

# Prior probabilities
labels = ['Before evidence\n(Prior)', 'After seeing FREE\n(Posterior)']
spam_probs = [0.4, round(p_spam_given_free, 2)]

plt.bar(labels, spam_probs, color=['blue', 'red'], alpha=0.7)
plt.title("Bayes Theorem — Spam Probability Update")
plt.ylabel("Probability of Spam")
plt.ylim(0, 1)

for i, v in enumerate(spam_probs):
    plt.text(i, v + 0.02, str(round(v*100, 1)) + "%",ha='center', fontweight='bold')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

scores = np.random.normal(loc=70, scale=10, size=1000)

plt.hist(scores, bins=30, color='blue', alpha=0.7)
plt.axvline(np.mean(scores), color='red', label='Mean')
plt.axvline(np.mean(scores) + np.std(scores), color='green', linestyle='--', label='Mean + SD')
plt.axvline(np.mean(scores) - np.std(scores), color='green', linestyle='--', label='Mean - SD')
plt.title("Score Distribution")
plt.legend()
plt.show()
```
